```python
# Analyses multiple IFC files and count the number of IfcDistributionPorts per model.
# Script failed with a model of 900mb, worked with models of serveral hundred mb.
# Version: 0.1 - 10.11.2025 - IFC4x3

import ifcopenshell
import os

# List of IFC file paths (replace with your actual file paths)
file_paths = [
    '/content/LU32_BS_MD_ME_FACHM.ifc',
    '/content/LU32_EL_MD_ME_FACHM.ifc',
    '/content/LU32_GA_MD_ME_FACHM.ifc',
    '/content/LU32_HK_MD_ME_FACHM.ifc',
    '/content/LU32_KG_MD_ME_FACHM.ifc',
    '/content/LU32_KU_MD_ME_FACHM.ifc',
    '/content/LU32_LG_MD_ME_FACHM.ifc',
    '/content/LU32_LI_MD_ME_FACHM.ifc',
    '/content/LU32_LU_MD_ME_FACHM.ifc',
    '/content/LU32_MG_MD_ME_FACHM.ifc',
    '/content/LU32_RP_MD_ME_FACHM.ifc',
    '/content/LU32_SA_MD_ME_FACHM.ifc',
    '/content/LU32_SP_MD_ME_FACHM.ifc',
    '/content/LU32_SR_MD_ME_FACHM.ifc',
    '/content/LU32_WL_MD_ME_WERKL.ifc'
]


total_count = 0
for file_path in file_paths:
    try:
        # Open the IFC file
        file = ifcopenshell.open(file_path)

        # Find all IfcDistributionPort elements
        distribution_ports = file.by_type('IfcDistributionPort')

        # Count the number of IfcDistributionPort elements
        count = len(distribution_ports)

        # Print the count and authoring software for the current file in a compact format
        print(f"File: {os.path.basename(file_path)}, Number of IfcDistributionPort elements: {count}")


        # Add to the total count
        total_count += count

    except Exception as e:
        print(f"Error processing {os.path.basename(file_path)}: {e}")

# Print the total count
print(f"\nTotal number of IfcDistributionPort elements across all models: {total_count}")
```
